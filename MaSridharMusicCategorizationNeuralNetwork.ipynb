```python
# import libraries
import tensorflow as tf
import numpy as np
import pandas as pd
import statistics
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('songs_normalize.csv')

xDf = df[['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
yDf = df['genre']
```

```python
y_relationship = np.reshape(yDf, (len(yDf), ))
for i, value in enumerate(yDf):
  x = value.split(", ")
  value = x[0]
  y_relationship[i] = value
print(y_relationship[0:15])
```

```python
# transforming genre data (y) into numerical data
for i, element in enumerate(yDf):
  if element == 'pop':
    yDf[i] = 1
  elif element == 'hip hop':
    yDf[i] = 2
  elif element == 'rock':
    yDf[i] = 3
  elif element == 'country':
    yDf[i] = 4
  elif element == 'metal':
    yDf[i] = 5
  elif element == 'R&B':
    yDf[i] = 6
  elif element == 'Dance/Electronics':
    yDf[i] = 7
  elif element == 'Folk/Acoustic':
    yDf[i] = 8
  elif element == 'latin':
    yDf[i] = 9
  elif element == 'blues':
    yDf[i] = 10
  else:
    yDf[i] = 0
print(yDf)
print(yDf.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(xDf, yDf, test_size = 0.2 )
x_train.shape
```

```python
# create a validation set
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
x_val.shape
```

```python
x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
y_val = y_val.to_numpy()
x_val = x_val.to_numpy()
```

```python
# create MLP structure
model = Sequential()
model.add(Input(shape=(10,)))
model.add(Dense(250, activation='relu'))
model.add(Dense(1, activation='softmax'))
```

```python
x_array = np.asarray(x_train).astype('float32')
x_val_array = np.asarray(x_val).astype('float32')
x_test_array = np.asarray(x_test).astype('float32')
y_array = np.asarray(y_train).astype('float32')
y_val_array = np.asarray(y_val).astype('float32')
y_test_array = np.asarray(y_test).astype('float32')
print(x_array.shape)
print(x_val_array.shape)
print(y_array.shape)
print(y_val_array.shape)
```

```python
# compile and train model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_array, y_array, epochs=10, batch_size=32, validation_data=(x_val_array, y_val_array))
```

```python
# evaluate accuracy
y_pred = model.predict(x_test)
y_pred = tf.argmax(y_pred, axis=1)
# y_test = tf.argmax(y_test_oh, axis=1)
loss, acc = model.evaluate(x_test_array, y_test_array)
print("Test accuracy:", acc)
print("Confusion matrix:\n", confusion_matrix(y_test_array, y_pred))
```
